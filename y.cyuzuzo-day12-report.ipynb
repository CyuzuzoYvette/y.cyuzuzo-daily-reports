```python

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
import seaborn as sns
import matplotlib.image as mpimg
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Create a simple line plot for y = x for x values ranging from 0 to 10.
# Exercise 1

x = np.linspace(0, 10, 10)
y = x ** 2

plt.figure(figsize=(7, 6))

plt.plot(x, y, 'b')

plt.xlabel('x')
plt.ylabel('y')

plt.title(r"Plot of $y=x^2$")
plt.show()
```

```python
#Plot y = x^2 using Matplotlib for x values between -10 and 10.
# Exercise 2

x = np.linspace(-10,10,1000)
y = x ** 2
plt.plot(x,y,color="blue")
plt.title(r"matplotlib for x")
plt.xlabel('x')
plt.ylabel('y')
```

```python
#Create a scatter plot with 50 random points where x and y values are between 0 and 10.
# Exercise 3

plt.figure(figsize=(10, 6), dpi=80)

n = 50
X = np.random.normal(0,10,n)
Y = np.random.normal(0,1,n)

plt.scatter(X,Y, s=45, c ="red")

# plt.xlim(-1.0, 1.0), plt.xticks([])
# plt.ylim(-1.0, 1.0), plt.yticks([])
plt.show()
```

Plot a sine wave using Matplotlib for x values between 0 and 2Ï€.

```python
#Plot a sine wave using Matplotlib for x values between 0 and 2Ï€.
# Exercise 4

X = np.linspace(0, 2*np.pi, 256, endpoint=True)
Y = np.sin(X)


plt.plot(X,Y, c = "green")

plt.title(r"Sine wave graph")
plt.xticks([0, np.pi/2,np.pi,3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])

plt.yticks([-1, 0, +1],
           [r'$-1$', r'$0$', r'$+1$'])

plt.show()
```

```python
#Create a bar chart for the following data: categories = ['A', 'B', 'C', 'D'] and values = [3, 7, 5, 9].
# Exercise 5

categories = ['A', 'B', 'C', 'D']  
values = [3, 7, 5, 9]  
plt.bar(categories, values, color='pink')  
plt.title('Bar Chart') 
plt.grid(axis='y')  
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
```

```python
#Create a histogram of 1000 random numbers generated from a normal distribution.
# Exercise 6


mean = 0    
std_dev = 1 
data = np.random.normal(mean, std_dev, 1000)  
plt.hist(data, bins=15, color='skyblue', edgecolor='black')  
plt.title('Histogram of Random Numbers from Normal Distribution')  
plt.xlabel('Value')  
plt.ylabel('Frequency')  
plt.show()
```

```python
#Plot two line plots on the same graph: y1 = x^2 and y2 = x^3 for x ranging from 0 to 10.
# Exercise 7

x = np.linspace(0, 10, 10)
y1 = x**2
y2 = x**3
plt.figure(figsize=(8, 6))

# plot values; with a color `blue`
plt.plot(x, y1, 'b')
plt.plot(x, y2, 'r')
plt.show()
```

```python
#Add titles, axis labels, and a legend to the plot in Exercise 7.
# Exercise 8


x = np.linspace(0, 10, 10)
y1 = x**2
y2 = x**3
plt.figure(figsize=(8, 6))

# plot values; with a color `blue`
plt.plot(x, y1, 'b',label="y = x**2")
plt.plot(x, y2, 'r',label="y = x**3")
plt.legend(loc=2);

# give a title to the plot
plt.title(r"Plot for $y1=x^2 and y1=x^3 $")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Create a subplot with 2 rows and 1 column. Plot y1 = x^2 in the first subplot and y2 = x^3 in the second.
# Exercise 9

x = np.linspace(-10, 10, 100)  

y1 = x**2   
y2 = x**3  
fig, axs = plt.subplots(2, 1, figsize=(8, 10)) 

# Plot y1 = x^2 in the first subplot  
axs[0].plot(x, y1, color='blue')  
axs[0].set_title('y = x^2')  
axs[0].set_xlabel('x')  
axs[0].set_ylabel('y')  
axs[0].grid(True)  

# Plot y2 = x^3 in the second subplot  
axs[1].plot(x, y2, color='orange')  
axs[1].set_title('y = x^3')  
axs[1].set_xlabel('x')  
axs[1].set_ylabel('y')  
axs[1].grid(True)  
plt.tight_layout()  
plt.show()
```

```python
#Use a different color and linestyle for each line in a plot of three functions: y1 = x, y2 = x^2, and y3 = x^3.
# Exercise 10

x = np.linspace(0, 10, 20)  
y1 = x  
y2 = x**2  
y3 = x**3   
plt.figure(figsize=(10, 6))   
plt.plot(x, y1, color='blue', linestyle='-', label='y = x')       
plt.plot(x, y2, color='green', linestyle='--', label='y = x^2')    
plt.plot(x, y3, color='red', linestyle=':', label='y = x^3')  
plt.title('Plots of y = x, y = x^2, and y = x^3')  
plt.xlabel('x')  
plt.ylabel('y')  
plt.legend()  
plt.grid()  
plt.show()
```

```python
#Create a pie chart with 5 segments labeled A, B, C, D, and E, with values [25, 35, 20, 15, 5].
# Exercise 11

labels = ['A', 'B', 'C', 'D', 'E']  
values = [25, 35, 20, 15, 5]  
# Create the pie chart  
plt.figure(figsize=(6, 6))  
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=30)  
 
plt.title('Pie Chart of Segments')  
plt.axis('equal')  

plt.show()
```

```python
#Plot a filled area under the curve of y = x^2 for x ranging from 0 to 10.
# Exercise 14

x = np.linspace(0, 10, 100)
y = x**2
plt.figure(figsize=(12, 8))
plt.plot(x, y, color="blue")
plt.fill_between(x, y, color="skyblue", alpha=0.4)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Change the style of the plot to ggplot and replot the sine wave from Exercise 4.
# Exercise 15

plt.style.use('ggplot')
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
plt.figure(figsize=(10, 6))
plt.plot(x, y, color="red")
plt.title('Sine Wave with ggplot Style')
plt.show()
```
